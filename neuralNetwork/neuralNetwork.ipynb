```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, LeakyReLU
from tensorflow.keras.activations import linear, relu, sigmoid
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy
```

```python
df_train = pd.read_csv("digitRecognizer/train.csv")
df_train = df_train.dropna()
df_train.describe()
```

```python
df_train.head(5)
```

```python
X = df_train.iloc[:,1:].to_numpy()
y = (df_train[["label"]]).to_numpy()

print("X shape: ", X.shape)
print("X: ", X[0:5,:])
print("y shape: ", y.shape)
print("y: ", y[0:5])
```

```python
tf.random.set_seed(1234)  # applied to achieve consistent results

initial_input_quantity = X.shape[0]

model = Sequential(
    [
    
        Dense(2000, activation='relu', name = 'layer1'),
        Dense(1000, activation='relu', name = 'layer2'),
        Dense(700, activation='relu', name = 'layer3'),
        Dense(500, activation='relu', name = 'layer4'),
        Dense(250, activation='relu', name = 'layer5'),
        Dense(125, activation='relu', name = 'layer6'),
        Dense(62, activation='relu', name = 'layer7'),
        Dense(10, activation = 'linear', name = "output")
     ]
)

model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),  #<-- Note
    optimizer=tf.keras.optimizers.Adam(0.001),
)

model.fit(
    X,y,
    epochs=10
)
```

```python
test_metrics = model.evaluate(X, y)
print("Metrics on test data:", test_metrics)
```

```python
predictions = model.predict(X)
sm_predictions = tf.nn.softmax(predictions).numpy()

m = X.shape[0]

predction_values = np.zeros(m)

for i in range(m):
    predction_values[i] = np.argmax(sm_predictions[i])

err = 0

for i in range(m):
    if(i%1000 == 0):
        print("Real label: %d" %y[i])
        print("Predction:  %d" %predction_values[i])
    if(y[i] != predction_values[i]):
        err+=1

print("Err: ", err)
```
